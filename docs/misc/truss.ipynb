## Linear bar element

Consider a bar element with length L, cross-sectional area A, and Young's modulus E. The bar is only subjected to axial forces, with the axial displacement as $u(x)$, where $x$ is the coordinate along the length of the bar. The axial strain can be expressed as: $\varepsilon(x) = \frac{du}{dx}$, the axial stress: $\sigma(x) = E \varepsilon(x) = E \frac{du}{dx}$, and the internal axial force: $N(x) = \sigma(x)A = E A \frac{du}{dx}$. Consider a distributed axial load $f(x)$ along the bar, the equilibrium equation can be expressed as: 
$$
\frac{d}{dx}(EA \frac{du}{dx}) + f(x) = 0, \quad 0 < x < L
$$
The boundary conditions can be specified at the ends of the bar, either as prescribed displacements or forces. For example, if the left end of the bar is fixed and the right end is subjected to a force $P$, the boundary conditions can be expressed as:
$$u(0) = 0, \quad EA \frac{du}{dx}\bigg|_{x=L} = P$$
Then we can derive the weak form of the equilibrium equation by multiplying the first equation with a test function $v(x)$ and integrating over the domain:
$$
\begin{align*}
&\int_0^L v(x) \frac{d}{dx}(EA \frac{du}{dx}) dx = -\int_0^L v(x) f(x) dx\nonumber\\
\end{align*}
$$
Applying integration by parts to the left-hand side, we have:
$$
\begin{align*}
&\int_0^L v(x) \frac{d}{dx}(EA \frac{du}{dx}) dx = \int_0^L EA \frac{du}{dx} \frac{dv}{dx} dx - [v(x) EA \frac{du}{dx}]_0^L\nonumber\\
\end{align*}
$$
Then we have:
$$
\begin{align*}
\int_0^L EA \frac{du}{dx} \frac{dv}{dx} dx = \int_0^L v(x) f(x) dx + [v(x) EA \frac{du}{dx}]_0^L
\end{align*}
$$
For the above boundary conditions, the term $[v(x) EA \frac{du}{dx}]_0^L$ can be simplified to $v(L) P$. Therefore, the weak form of the equilibrium equation can be expressed as:
$$
\int_0^L EA \frac{du}{dx} \frac{dv}{dx} dx = \int_0^L v(x) f(x) dx + v(L) P
$$

Alternatively, we can derive the finite element formulation from the energy perspective. The strain energy stored in the bar element can be expressed as:
$$
U_e = \frac{1}{2} \int_0^L EA \left(\frac{du}{dx}\right)^2 dx = \frac{1}{2} \int_0^L EA \left(\frac{u_2-u_1}{L}\right)^2 dx = \frac{1}{2} \frac{EA}{L} (u_2-u_1)^2
$$
The internal force can be derived from the strain energy as:
$$
\boldsymbol{f}_e = \frac{\partial U_e}{\partial \boldsymbol{u}} = \frac{EA}{L} \begin{bmatrix} -1 \\ 1 \end{bmatrix} (u_2-u_1)
$$
The stiffness matrix can be derived from the internal force as:
$$
K_e = \frac{\partial \boldsymbol{f}_e}{\partial \bold{u}} = \frac{EA}{L} \begin{bmatrix} 1 & -1 \\ -1 & 1 \end{bmatrix}
$$
